```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# sphinx_gallery_thumbnail_number = 2

crop = ["","soy", "wheat","corn",""]
year = ["2016","2017","2018","2019"]

#left this the same, I did not feel like converting bu/ac for every field to total harvest.
#added "buffe rows because it looks slightly better
harvest = np.array([[0, 0, 0,0 ],
                    [0.8, 2.4,  4.0, 3.0],
                    [1.3, 1.2, 3.2, 5.1],
                    [0.1, 2.0, 1.9, 6.3],
                    [0, 0, 0,0 ]])


fig, ax = plt.subplots()
im = ax.imshow(harvest)

# We want to show all ticks...
ax.set_xticks(np.arange(len(year)))
ax.set_yticks(np.arange(len(crop)))
# ... and label them with the respective list entries
ax.set_xticklabels(year)
ax.set_yticklabels(crop)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(crop)):
    for j in range(len(year)):
        text = ax.text(j, i, harvest[i, j],
                       ha="center", va="center", color="w")

ax.set_title("Harvest from 2016 through 2019 (in tons/year)")
fig.tight_layout()
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


labels = ['2016', '2017', '2018', '2019']
harvest_c = [2, 3, 3, 4]
harvest_s = [5, 2, 4, 2]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, harvest_c, width, label='Corn')
rects2 = ax.bar(x + width/2, harvest_s, width, label='Soy')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Tonnes')
ax.set_title('Harvest (in tonnes) from 2016 through 2019')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
